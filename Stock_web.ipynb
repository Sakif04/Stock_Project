# Import Stock Data from Dhaka Stock Exchange website

```python
import pandas as pd
import matplotlib.pyplot as plt
stock=pd.read_html('https://www.amarstock.com/latest-share-price',index_col='Trading code')
stock_df=stock[0]
stock_df
```

## Get the stock datas with the positive changes only

```python
positive_stock=stock_df[stock_df['Change %']>0][['Change %','Change','Trade']]
positive_stock.plot(figsize=(28,12),kind='bar',y='Change %' )
```

# Get the Stock data with Negative Changes of selected column


```python
negative_stock=stock_df[stock_df['Change %']<0][['Trade','Change %','Change']]
neg_plot=negative_stock.plot(figsize=(155,18),y='Change %',kind='bar')
```

## Get top sorted 25% Traded Stocks from the dataset with Positive changes and 


```python
top_traded=stock_df[(stock_df['Trade']>stock_df['Trade'].quantile(0.75)) & (stock_df['Change']>0)][['Trade','Change %','LTP']]
top_traded.sort_values(by=['Trade'],inplace=True)
top_traded.plot(kind='bar',y='Trade',title='Trades')
top_traded.plot(kind='bar',y='LTP',title='Price')
```

# Get stocks with price lower than 10 and plot them 

```python
low_priced_stock=stock_df[stock_df['LTP']<10]
low_priced_stock.plot(kind='bar',y='LTP',figsize=(25,10))
```
